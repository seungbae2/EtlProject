```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
from datetime import datetime
```

```python
player_filepath = "Resources/basketball_player_allstar.csv"
player_df = pd.read_csv(player_filepath)
player_df.head()
```

```python
game_filepath = "Resources/NBA_All_Star_Games.xlsx"
game_df = pd.read_excel(game_filepath)
game_df.head()
```

```python
player_df = player_df.loc[player_df['season_id'] >= 2000]
player_df['full_name'] = player_df['first_name'] + " " +  player_df['last_name']
player_df['full_name']
player_df.head()
```

```python
player_df.dropna()
player_df.head()
```

```python
player_df.columns
```

```python
columns_player = ['player_id', 'last_name', 'first_name', 'year', 'conference',
       'league_id', 'games_played', 'minutes', 'points', 'offensive_rebounds',
       'defensive_rebounds', 'total_rebounds', 'assists', 'steals', 'blocks', 'turnovers',
       'personal_fouls', 'field_goals_attempted', 'field_goals_made', 'free_throws_attempted', 'free_throw_made',
       'three_points_attempted', 'three_points_made', 'player']
player_df.columns = columns_player
player_df.head()
```

```python
del player_df['player_id']
del player_df['last_name']
del player_df['first_name']
player_df.head()
```

```python
game_df = game_df.loc[game_df['Year'] <= 2009]
```

```python
game_df
```

```python
## Height Dateobj to string
height = []

for game in game_df['HT']:
    if not isinstance(game, str):
        height.append(game.strftime("%m-%d").replace("0", ""))
    else:
        height.append(game)
```

```python
game_df.insert(3, 'height', height)
del game_df['HT']
```

```python
game_df.columns
```

```python
columns_game = ['year', 'player', 'position', 'height', 'weight', 'team', 'selection',
       'nba_draft_status', 'home_country']
game_df.columns = columns_game
game_df.head()
```

```python
engine = create_engine('postgresql://postgres:'+ password +'@localhost:5432/NbaAllstar')
connection = engine.connect()
# Confirm tables
engine.table_names()
```

```python
player_df.to_sql(name='allstar_players', con=engine, if_exists='replace', index=False)
```

```python
game_df.to_sql(name='allstar_games', con=engine, if_exists='replace', index=False)
```
